# Companion notebook for SI extrinsic noise fraction figure

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def convert_xy_to_params(x, y, beta, gamma, K_avg):
    kappa = (beta + gamma)*(x/(1-x))
    a_over_th = 1/y - 1
    theta = np.sqrt(kappa*K_avg/a_over_th)
    a = K_avg*kappa/theta
    return a, kappa, theta


def plot_ext_fraction_slowrev(ax, x, show_label=True):
    
    ext_frac = 1/(1 + x)
    
  
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.set_xticks([0, 5, 10, 15, 20])
    ax.set_yticks([0, 0.25, 0.5, 0.75, 1])
    ax.set_yticklabels(['0', 0.25, 0.5, 0.75, '1'])
    ax.tick_params(axis='both', which='major', labelsize=20)
    
    ax.plot(x, ext_frac, color='darkorange', linewidth=4)
    ax.plot(x, np.ones_like(x), color='black', linestyle='--', linewidth=2)
    ax.set_xlim(-0.5,None)
    ax.set_ylim(0,1.1)
    if show_label:
        ax.set_xlabel("$\\alpha/\\mu_N$", fontsize=20)
        ax.set_ylabel("extrinsic fraction", fontsize=20)
        
    return

def plot_ext_fraction_highgain(ax, x, show_label=True):
    
    ext_frac = x/(1 + x)
    
  
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.set_xticks([0, 5, 10, 15, 20])
    ax.set_yticks([0, 0.25, 0.5, 0.75, 1])
    ax.set_yticklabels(['0', 0.25, 0.5, 0.75, '1'])
    ax.tick_params(axis='both', which='major', labelsize=20)
    
    ax.plot(x, ext_frac, color='darkorange', linewidth=4)
    ax.plot(x, np.ones_like(x), color='black', linestyle='--', linewidth=2)
    ax.set_xlim(-0.5,None)
    ax.set_ylim(0, 1.1)
    if show_label:
        ax.set_xlabel("$b$", fontsize=20)
        ax.set_ylabel("extrinsic fraction", fontsize=20)

    return


def plot_ext_fraction_vs_y(ax, x, y, mu_N, show_label=True):
    

    arg = np.sqrt(mu_N * x * (1-x) * y / (1 - y))
    ext_frac = arg/(arg + 1)
    
    arg2 = np.sqrt(1 * x * (1-x) * y / (1 - y))
    ext_frac2 = arg2/(arg2 + 1)
    
    arg3 = np.sqrt(100 * x * (1-x) * y / (1 - y))
    ext_frac3 = arg3/(arg3 + 1)
    
  
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.set_xticks([0, 0.25, 0.5, 0.75, 1])
    ax.set_yticks([0, 0.25, 0.5, 0.75, 1])
    ax.set_xticklabels([0, 0.25, 0.5, 0.75, '1'])
    ax.set_yticklabels(['0', 0.25, 0.5, 0.75, '1'])
    ax.tick_params(axis='both', which='major', labelsize=20)
    
    ax.plot(y, ext_frac, color='darkorange', linewidth=4)
    ax.plot(y, ext_frac2, color='orange', linewidth=4)
    ax.plot(y, ext_frac3, color='chocolate', linewidth=4)
    ax.plot(y, np.ones_like(y), color='black', linestyle='--', linewidth=2)
    ax.set_xlim(-0.1,None)
    ax.set_ylim(0, 1.1)
    if show_label:
        ax.set_xlabel("$y$", fontsize=20)
        ax.set_ylabel("extrinsic fraction", fontsize=20)
    
    return


def plot_ext_fraction_vs_x(ax, x, y, mu_N, show_label=True):
    

    arg = np.sqrt(mu_N * x * (1-x) * y / (1 - y))
    ext_frac = arg/(arg + 1)
    
    arg2 = np.sqrt(1 * x * (1-x) * y / (1 - y))
    ext_frac2 = arg2/(arg2 + 1)
    
    arg3 = np.sqrt(100 * x * (1-x) * y / (1 - y))
    ext_frac3 = arg3/(arg3 + 1)
    
  
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.set_xticks([0, 0.25, 0.5, 0.75, 1])
    ax.set_yticks([0, 0.25, 0.5, 0.75, 1])
    ax.set_xticklabels([0, 0.25, 0.5, 0.75, '1'])
    ax.set_yticklabels(['0', 0.25, 0.5, 0.75, '1'])
    ax.tick_params(axis='both', which='major', labelsize=20)
    
    ax.plot(x, ext_frac, color='darkorange', linewidth=4)
    ax.plot(x, ext_frac2, color='orange', linewidth=4)
    ax.plot(x, ext_frac3, color='chocolate', linewidth=4)
    ax.plot(x, np.ones_like(x), color='black', linestyle='--', linewidth=2)
    ax.set_xlim(-0.1,None)
    ax.set_ylim(0, 1.1)
    if show_label:
        ax.set_xlabel("$x$", fontsize=20)
        ax.set_ylabel("extrinsic fraction", fontsize=20)

    return



def figure_row1():
    
    fig, ax = plt.subplots(1,4, figsize=((24,4)))
    
    num_x = 100
    x = np.linspace(0, 20, num_x)
    plot_ext_fraction_slowrev(ax[0], x, show_label=False)
    plot_ext_fraction_highgain(ax[1], x, show_label=False)
    
    mu_N = 10
    
    y = 0.99
    num_x = 1000
    x = np.linspace(0, 1, num_x)
    plot_ext_fraction_vs_x(ax[2], x, y, mu_N, show_label=False)
    
    
    x = 0.5
    num_y = 1000
    epsilon = 1e-6
    y = np.linspace(0, 1-epsilon, num_y)
    plot_ext_fraction_vs_y(ax[3], x, y, mu_N, show_label=False)


    plt.subplots_adjust(wspace=0.3)
    name = 'ext_frac_SI_row1'
    plt.savefig('results/'+name+'.png', bbox_inches='tight')
    plt.savefig('results/'+name+'.svg', bbox_inches='tight')
    plt.show()
    return



def figure_row2():
    
    beta, gamma = 1, 0
    
    # Set up qualitative regimes grid
    x_min, x_max, num_x = 0.01, 0.99, 100
    y_min, y_max, num_y = 0.01, 0.99, 100
    x, y = np.linspace(x_min, x_max, num_x), np.linspace(y_min, y_max, num_y)
    X, Y = np.meshgrid(x, y, indexing='ij')
    
    ##
    
    fig, ax = plt.subplots(1,4, figsize=((18,4)), gridspec_kw={'width_ratios':[1, 1, 1, 0.3]})
    
    
    K_avg = 1
    a, kappa, theta =  convert_xy_to_params(X, Y, beta, gamma, K_avg)
    extrinsic_fraction = (theta)/(theta + beta + kappa)
    
    pos = ax[0].imshow(np.transpose(extrinsic_fraction), origin='lower', extent=[x_min, x_max, y_min, y_max], vmin=0, vmax=1)
    ax[0].set_xticks([0.2, 0.4, 0.6, 0.8])
    ax[0].set_yticks([0.2, 0.4, 0.6, 0.8])
    ax[0].tick_params(axis='both', which='major', labelsize=15)
    
    ##
    
    K_avg = 10
    a, kappa, theta =  convert_xy_to_params(X, Y, beta, gamma, K_avg)
    extrinsic_fraction = (theta)/(theta + beta + kappa)
    
    pos = ax[1].imshow(np.transpose(extrinsic_fraction), origin='lower', extent=[x_min, x_max, y_min, y_max], vmin=0, vmax=1)
    ax[1].set_xticks([0.2, 0.4, 0.6, 0.8])
    ax[1].set_yticks([0.2, 0.4, 0.6, 0.8])
    ax[1].tick_params(axis='both', which='major', labelsize=15)

    ##
    
    K_avg = 100
    a, kappa, theta =  convert_xy_to_params(X, Y, beta, gamma, K_avg)
    extrinsic_fraction = (theta)/(theta + beta + kappa)
    
    pos = ax[2].imshow(np.transpose(extrinsic_fraction), origin='lower', extent=[x_min, x_max, y_min, y_max], vmin=0, vmax=1)
    ax[2].set_xticks([0.2, 0.4, 0.6, 0.8])
    ax[2].set_yticks([0.2, 0.4, 0.6, 0.8])
    ax[2].tick_params(axis='both', which='major', labelsize=15)
    
    fake = np.array([[0,1]])
    img = ax[3].imshow(fake)
    ax[3].set_visible(False)
    ax[3].axis('off')
    cbar = fig.colorbar(img, ax=ax[3], ticks=[0, 0.25, 0.5, 0.75, 1], orientation="vertical")
    ax[3].yaxis.set_label_position('left')
    cbar.ax.set_yticklabels([0, 0.25, 0.5, 0.75, 1])
    cbar.ax.tick_params(labelsize=15)
    ax[3].tick_params(axis='both', which='major', labelsize=15)
    
    ## 
    plt.subplots_adjust(wspace=-0.6)
    
    name = 'ext_frac_SI_row2'
    plt.savefig('results/'+name+'.png', bbox_inches='tight')
    plt.savefig('results/'+name+'.svg', bbox_inches='tight')
    plt.show()
    return
```

```python
figure_row1()
figure_row2()
```
